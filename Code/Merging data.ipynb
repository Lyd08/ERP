### Making % columns for communities

```python
acs = lookup.copy()

# Total households for income %
income_cols = ['Under $25,000', '$25,000 to $49,999',
               '$50,000 to $74,999', '$75,000 to $125,000', '$125,000 +']
acs['total_households'] = acs[income_cols].sum(axis=1)

# % Low income (<$50k)
acs['pct_low_income'] = ((acs['Under $25,000'] + acs['$25,000 to $49,999']) /
                         acs['total_households']) * 100

# % Black, Hispanic, Nonwhite
acs['pct_black'] = (acs['Black or African American'] / acs['Total Population']) * 100
acs['pct_hispanic'] = (acs['Hispanic or Latino'] / acs['Total Population']) * 100
acs['pct_nonwhite'] = 100 - (acs['White Not Hispanic or Latino'] /
                             acs['Total Population']) * 100
```

#### Drop-off merging

```python
# Merge for dropoff if needed
# Merge ACS on dropoff
trips = trips.merge(acs[['Community Area Number', 'pct_low_income', 'pct_black', 'pct_hispanic', 'pct_nonwhite','HARDSHIP INDEX']],
                    how='left',
                    left_on='Dropoff Community Area',
                    right_on='Community Area Number',
                    suffixes=('', '_drop'))

# Rename drop-off features
trips.rename(columns={
    'pct_low_income_drop': 'dropoff_pct_low_income',
    'pct_black_drop': 'dropoff_pct_black',
    'pct_hispanic_drop': 'dropoff_pct_hispanic',
    'pct_nonwhite_drop': 'dropoff_pct_nonwhite',
    'HARDSHIP INDEX': 'dropoff_Hardship_index'
}, inplace=True)

```

```python
trips.rename(columns={
    'HARDSHIP INDEX_drop': 'dropoff_Hardship_index',
    'dropoff_Hardship_index': 'pickup_Hardship_index'
}, inplace=True)
```

```python
trips.drop(columns=['Community Area Number','Community Area Number_drop'], inplace=True)
```

```python
trips["Pickup Community Area"] = trips["Pickup Community Area"].astype("Int32")
trips["Dropoff Community Area"] = trips["Dropoff Community Area"].astype("Int32")
```

```python
# Parse datetime with specified format
trips['Trip Start Timestamp'] = pd.to_datetime(
    trips['Trip Start Timestamp'],
    format="%m/%d/%Y %I:%M:%S %p",
    errors='coerce'
)
trips['Trip End Timestamp'] = pd.to_datetime(
    trips['Trip End Timestamp'],
    format="%m/%d/%Y %I:%M:%S %p",
    errors='coerce'
)
trips['start_hour'] = trips['Trip Start Timestamp'].dt.hour
trips['day_of_week'] = trips['Trip Start Timestamp'].dt.day_name()
```
